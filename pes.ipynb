```python
import altair as alt
import pint
import polars as pl

eh2kcal = pint.Quantity("hartree/molecule").m_as("kcal/mol")

data = [
    {"name": "RH+OH", "raw_energy": -195.006948667186 + -75.652016292465},
    {"name": "RH+OH-c1", "raw_energy": -270.665551608314},
    {"name": "RH+OH-c2", "raw_energy": -270.664841679987},
    {"name": "RH+OH-c_RHOH_saddle", "raw_energy": -270.663383567243},
    {"name": "RHOH", "raw_energy": -270.710776742582},
]
df = pl.DataFrame(data).with_row_index()
df = df.with_columns(energy=(pl.col("raw_energy") - pl.first("raw_energy")) * eh2kcal)

points = (
    alt.Chart(df)
    .mark_point()
    .encode(x="index:Q", y=alt.Y("energy:Q").scale(zero=False))
)

text = points.mark_text(align="center", baseline="top", dy=12).encode(text="name")

points + text
```
